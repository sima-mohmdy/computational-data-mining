# find min euclidean distance between two vectors of test data & means of each classes in train data

```python
predicted_labels = []

for img in test:
  # Flattening image in test data
  flattened_img = img.flatten()

  distance = []
  for m in mean_list:
    # Flattening mean of train data
    flattened_mean = m.flatten()

    # Calulate Euclidean distance
    d = np.linalg.norm(flattened_img - flattened_mean)

    distance.append(d)

  predicted_labels.append(np.argmin(distance))


#print('predicted labels is: ', predicted_labels)
print('predicted labels is: ', len(predicted_labels))
```
