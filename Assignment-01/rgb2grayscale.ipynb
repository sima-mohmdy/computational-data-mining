# Convert RGB to grayscale image without library

```python
import matplotlib.image as mpimg

def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray

img = mpimg.imread('/content/drive/MyDrive/ML/computational data mining/تمرین اول/test.png')
gray = rgb2gray(img)
plt.imshow(gray, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```
